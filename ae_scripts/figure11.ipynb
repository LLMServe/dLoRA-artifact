```python
%matplotlib inline
import os
import sys
import re
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import csv

# constants
num_subfigs = 1
num_curves = 3

rate_dict = {0: [], 1: [], 2: []}
queuing_delay_dict = {0: [], 1: [], 2: []}

file_path = "./fig11a.csv"
with open(file_path, 'r') as csvfile:
    reader = csv.DictReader(csvfile)

    for row in reader:
        print(row)
        migration_type = int(row["migration_type"]) - 1
        rate = float(row["rate"])
        queuing_delay = float(row["queueing_delay"])
        rate_dict[migration_type].append(rate)
        queuing_delay_dict[migration_type].append(queuing_delay)

# Set font and figure size
font_size = 48
plt.rc('font',**{'size': font_size})
plt.rc('pdf',fonttype = 42)

fig_size = (14, 10)
fig, axes = plt.subplots(nrows=1, ncols=1, sharey=False, figsize=fig_size)
matplotlib.rcParams['xtick.minor.size'] = 4.
matplotlib.rcParams['xtick.major.size'] = 8.
matplotlib.rcParams['ytick.major.size'] = 6.
matplotlib.rcParams['ytick.minor.visible'] = False
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.3, hspace=None)

colors = {0: 'green', 1: 'dodgerblue', 2: 'darkorange'}
labels = {0: 'RR',  1: 'Proactive', 2: 'dLoRA'}
markers = {0: 'o', 1: 's', 2: '^'}
linestyles = {0: 'solid', 1: 'solid', 2: '--'}

# x-axis setting
x_label = 'Arrival Rate (reqs/s)'
x_ticks = [i for i in range(0, 36, 4)]

# y-axis setting
y_label = 'Queuing Delay (s/token)'
y_ticks = [0, 0.4, 0.8, 1.2]

axes.set_xlabel(x_label)
axes.set_xlim(left=0.5, right=6)
axes.get_xaxis().set_tick_params(direction='in', pad=7)
axes.get_xaxis().set_tick_params(which='minor', direction='in')
axes.set_xticks(x_ticks)

axes.set_ylabel(y_label)
axes.set_ylim(bottom=0, top=1)
axes.set_yticks(y_ticks)
axes.get_yaxis().set_tick_params(direction='in', pad=4)
axes.get_xaxis().set_tick_params(direction='in', pad=4)
axes.tick_params(bottom=False, top=False, left=False, right=False)
axes.minorticks_off()

lines = [[] for i in range(num_curves * 1)]
for i in range(num_curves):
    lines[i], = axes.plot(
        rate_dict[i],
        queuing_delay_dict[i],
        label=labels[i],
        marker = markers[i], 
        color=colors[i],
        lw=5,
        markersize=10,
        linestyle=linestyles[i],
        zorder=3
    )

fig.legend(handles=[lines[0], lines[1], lines[2]], handlelength=1.1, 
           ncol=3, loc='upper center', bbox_to_anchor=(0.5, 1.01), frameon=False, prop={'size':font_size})

file_path = './figure11a.pdf'
plt.savefig(file_path, bbox_inches='tight', transparent=True)
```

```python
%matplotlib inline
import os
import sys
import re
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# constants
num_size = 5
num_bars = 3
num_subfigs = 1

ratio_dict = {0: [], 1: [], 2: []}
avg_latency_dict = {0: [], 1: [], 2: []}

file_path = "./fig11b.csv"
with open(file_path, 'r') as csvfile:
    reader = csv.DictReader(csvfile)

    for row in reader:
        print(row)
        migration_type = int(row["migration_type"]) - 1
        ratio = int(row["ratio"])
        avg_latency = float(row["avg_lat"])
        ratio_dict[migration_type].append(ratio)
        avg_latency_dict[migration_type].append(avg_latency)

# Set font and figure size
font_size = 48
plt.rc('font',**{'size': font_size, 'family': 'Arial'})
plt.rc('pdf',fonttype = 42)

fig_size = (14, 9.3)
fig, axes = plt.subplots(figsize=fig_size)
matplotlib.rcParams['xtick.minor.size'] = 4.
matplotlib.rcParams['xtick.major.size'] = 8.
matplotlib.rcParams['ytick.major.size'] = 6.
matplotlib.rcParams['ytick.minor.visible'] = True
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.3, hspace=None)

# settings
colors = {0: 'green', 1: 'dodgerblue', 2: 'darkorange'}
labels = {0: 'RR', 1: 'Proactive', 2: 'dLoRA'}

width = 0.05
indexes = [[0.2*i+0.1+width*j for i in range(num_size)] for j in range(num_bars)]

# x-axis setting
x_label = 'Skewness (stride)'
x_ticks = []
for i in range(num_size):
    x_ticks.append(sum(indexes[j][i] for j in range(num_bars))/num_bars)
x_ticklabels = ['1', '2', '4', '6', '8']

# y-axis setting
y_label = 'P90 Latency (s/token)'
y_ticks = np.arange(0, 40, step=5)

axes.set_xlabel(x_label)
axes.set_xticks(x_ticks)
axes.set_xticklabels(x_ticklabels)
axes.get_xaxis().set_tick_params(direction='in', pad=7)
axes.get_xaxis().set_tick_params(which='minor', direction='in')

axes.set_ylabel(y_label)
axes.set_yticks(y_ticks)
axes.yaxis.set_ticks_position('left')
axes.tick_params(bottom=False, top=False, left=False, right=False)
axes.minorticks_off()
axes.get_yaxis().set_tick_params(direction='in', pad=4)
axes.get_yaxis().set_tick_params(which='minor', direction='in')

# Plot bars
bars = [[] for i in range(num_bars)]
for i in range(num_bars):
    bars[i] = axes.bar(indexes[i], avg_latency_dict[i], width, label=labels[i], edgecolor='black', zorder=3, color = colors[i])


fig.legend(handles=[bars[0], bars[1], bars[2]], handlelength=1.1, 
           ncol=3, loc='upper center', bbox_to_anchor=(0.5, 1.01), frameon=False, prop={'size':font_size}, columnspacing = 1)

file_path = './figure11b.pdf'
plt.savefig(file_path, bbox_inches='tight', transparent=True)
```
